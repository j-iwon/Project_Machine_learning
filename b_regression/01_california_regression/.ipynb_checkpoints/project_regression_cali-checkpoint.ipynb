### **캘리포니아 집값**
> **longitude** 경도  
> **latitude** 위도  
> **housing_median_age** 블록 주택의 평균 연식  
> **total_rooms** 블록에 있는 주택의 총 방 수 (침실 제외)  
> **total_bedrooms** 블록에 있는 주택의 총 침실 개수  
> **population** 블록에 있는 총 인 수  
> **households** 블록에 있는 총 가구 수  
> **median_income** 블록에 있는 주택의 가구 소득의 중앙값   
> **ocean_proximity** 블록의 경관 유형   
> **median_house_value** 블록 내 가격의 중앙값  

```python
import pandas as pd

c_df = pd.read_csv('./datasets/1553768847-housing.csv')
c_df
```

#### **데이터 확인**

```python
c_df.info()
```

```python
# 결측치 확인
c_df.isna().sum()
```

```python
# 결측치 제거
c_df = c_df[~ c_df['total_bedrooms'].isna()]
c_df.isna().sum()
```

```python
# 데이터 중복행 확인
c_df.duplicated().sum()
```

#### **데이터 전처리**

```python
# 데이터 복사
pre_c_df = c_df.copy()
```

```python
# 범주형 데이터 확인
pre_c_df['ocean_proximity'].value_counts()
```

> <1H OCEAN : 1시간 이내 바다와 근접성  
INLAND : 바다와 떨어진 내륙 지역  
NEAR OCEAN : 바다와 가까운 지역  
NEAR BAY : 작은 만 근처에 위치한 지역  
ISLAND : 섬 지역  

```python
# 범주형 데이터 정리
pre_c_df['ocean_proximity'] = pre_c_df['ocean_proximity'].replace({
    '<1H OCEAN': 'OCEAN',
    'NEAR OCEAN': 'NEAR OCEAN',
    'NEAR BAY': 'NEAR OCEAN'
})
pre_c_df['ocean_proximity'].value_counts()
```

```python
# 섬 지역 데이터 제거하기
pre_c_df = pre_c_df.drop(pre_c_df[pre_c_df['ocean_proximity'] == 'ISLAND'].index)
pre_c_df['ocean_proximity'].value_counts()
```

```python
# 수치형 데이터 프레임 생성
columns = ['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','median_house_value']
num_p_df = pre_c_df[columns]
num_p_df
```

```python
num_p_df.hist(figsize=(10,8))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = num_p_df.iloc[:, :-1], num_p_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

#### **1Cycle 결과**
> 경도와 위도는 수치로 나타나지만 위치를 나타내는 범주형 데이터이다.  
> 경도 위도 컬럼을 제거해본다.  

```python
p_2c_df = num_p_df.drop(labels=['longitude', 'latitude'], axis=1).reset_index(drop=True)
p_2c_df
```

```python
(X_test.value_counts() <= 0).sum()
```

#### **2Cycle 결과**
> 전보다 R2는 하락했지만 범주형 데이터를 삭제한 결과가 맞다고 판단한다.  
> R2 수치가 나쁘지 않으므로 트리를 사용해 분석해본다.   

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

features, targets = p_2c_df.iloc[:, :-1], p_2c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor()
lgb_r = LGBMRegressor(n_estimators=100)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]
for model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print(model.__class__.__name__)
    get_evaluation(y_test, prediction)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

#### **5Cycle 결과**
> 타겟 분포 조정 결과 R2 결과가 전체적으로 상승하였다.   
> 최종적으로 LGBM 모델을 사용하는 것이 제일 적합하다 판단된다.  

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

features, targets = p_3c_df.iloc[:, :-1], p_3c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

lgb_r = LGBMRegressor()

parameters = {
    'n_estimators': [50, 100, 500],
    'learning_rate': [0.1 ,0.3, 0.5, 0.7, 1]
}
kfold = KFold(n_splits=10, random_state=124, shuffle=True)

# grid_lgb_r = GridSearchCV(lgb_r, param_grid=parameters, scoring='neg_mean_squared_error', cv=kfold)
grid_lgb_r = GridSearchCV(lgb_r, param_grid=parameters, scoring='r2', cv=kfold)
grid_lgb_r.fit(X_train, y_train)
```

```python
result_df = pd.DataFrame(grid_lgb_r.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

> 피쳐엔지니어링을 통해 가중치를 더한 독립변수와의 결과와, 새로운 피쳐를 생성하여 결과를 비교하였을 때 전이 훨씬 더 나은 결과를 가져왔다.
